```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
#accuracy値でcsvファイル書き込み
#二分前の(前後4分、6分で変化可能なように)距離と比較した停止判定
import csv

min_distance = 0#停止判定する閾値
max_accuracy = 0

input_file = open('route.csv', 'r')
output_file = open('route_modified.csv','w')

csvWriter = csv.writer(output_file)
before_latitude = 0#一つ前の緯度
before_longtitude = 0#一つ前の経度

min_distance = 200
max_accuracy = 500

def distance_check(min_distance,line):#min_distanceよりでかければtrue
    if ((((before_latitude - float(line.split(",")[1]))*111263.283)**2.0) + (((before_longtitude - float(line.split(",")[2]))*91000.0)**2.0))**0.5 > min_distance:
        return True
    else:
        return False

def accuracy_check(max_accuracy,line):#max_accuracyより小さければtrue
    if float(line.split(",")[4]) < max_accuracy:
        return True
    else:
        return False

for line in input_file:
    line = line.replace(line.split(",")[5],"")
    if distance_check(min_distance,line) and accuracy_check(max_accuracy,line):
        csvWriter.writerow([line])
    before_latitude = float(line.split(",")[1])
    before_longtitude = float(line.split(",")[2])
    #if float(line.split(",")[4]) < 500.0:#accuracy500で切り捨て
     #   csvWriter.writerow([line])
          
input_file.close()
output_file.close()
```

```python
#!/usr/bin/env python

f = open('route_modified.csv', 'r')
time = []
latitude = []#緯度
longtitude = []#経度
gpstype = []
accuracy = []

for line in f:
    time.append(line.split(",")[0])
    latitude.append(float(line.split(",")[1]))
    longtitude.append(float(line.split(",")[2]))
    gpstype.append(line.split(",")[3])
    accuracy.append(float(line.split(",")[4]))
    
f.close()

x = latitude
y = longtitude
plt.title('where am I')
plt.plot(x,y)

x = latitude
y = longtitude
z = accuracy
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x,y,z)
```

```python
distance = []#二分間の移動距離(m)
distance.append(0)

for num in range(1,len(time)):
    distance.append((((latitude[num - 1] - latitude[num])*111263.283)**2.0 + ((longtitude[num - 1] - longtitude[num])*91000.0)**2.0)**0.5)

x = distance
y = accuracy
plt.plot(x,y,".")
```
