```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
#!/usr/bin/env python

f = open('1213route.csv', 'r')
time = []
latitude = []#緯度
longtitude = []#経度
gpstype = []
accuracy = []

for line in f:
    time.append(line.split(",")[0])
    latitude.append(float(line.split(",")[1]))
    longtitude.append(float(line.split(",")[2]))
    gpstype.append(line.split(",")[3])
    accuracy.append(float(line.split(",")[4]))
    
f.close()

x = latitude
y = longtitude
plt.title('where am I')
plt.plot(x,y)

x = latitude
y = longtitude
z = accuracy
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x,y,z)
```

```python
distance = []#二分間の移動距離
distance.append(0)

for num in range(1,len(time)):
    distance.append(((latitude[num - 1] - latitude[num])**2.0 + (longtitude[num - 1] - longtitude[num])**2.0)**0.5)

x = distance
y = accuracy
plt.plot(x,y,".")
```
